```python
import tushare as ts 
df = ts.get_k_data('399300', index = True, start = '2016-01-01', end = '2016-12-31')
df.head()
```

```python
import numpy as np 
df['rtn'] = np.log(df['close']) - np.log(df['close'].shift(1))
df = df.dropna()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df['close'].plot()
plt.show()

df['rtn'].plot()
plt.show()
```
