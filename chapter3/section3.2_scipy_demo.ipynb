```python
from scipy import optimize
import matplotlib.pyplot as plt  

def f(x):
    return x**2 + 20*np.sin(x)

x = np.arange(-10, 10, 0.1)

plt.plot(x, f(x))
```

```python
grid = (-10, 10, 0.1)
x_min = optimize.brute(f, (grid, ))
x_min
```
